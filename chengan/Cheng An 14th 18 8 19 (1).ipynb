```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = load_boston()

df = pd.DataFrame(data.data,columns=data.feature_names)
df.head(40)
```

```python
df.head(10)

x = df[['CRIM','TAX']]
z = pd.DataFrame(data.target,columns=['House Price'])
```

```python
x.head()
```

```python
z.head()
```

```python
from sklearn.linear_model import LinearRegression
model_linear = LinearRegression()

model_linear.fit(x,z)

print(model_linear.intercept_,(model_linear.coef_))
```

```python
x1 = np.linspace(min(x['CRIM']),max(x['CRIM']),40)
x2 = np.linspace(min(x['TAX']),max(x['TAX']),40)

x1x2 = np.c_[x1,x2]
zz = model_linear.predict(x1x2)

print(x1x2)
```

```python
zz.reshape(1,-1)[0]
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter3D(x['CRIM'],x['TAX'],z,s=2)
ax.grid(True)
ax.plot3D(x1,x2,zz.reshape(1,-1)[0],'r--',lw=5)
```

```python
model_linear.score(x, z)
```
