```python
import scipy.stats as st
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("traffic-collision-data-from-2010-to-present.csv",sep=',')
```

```python
abbrev_data = data.head(200)
abbrev_data
```

```python
time = abbrev_data["Time Occurred"]
```

```python
print(f"Low: {time.min()}\nHigh: {time.max()}")
time_range = np.arange(0000,2500,100)

# Time ranges for plotting
print(time_range)
```

```python
freq_list = []

for count in range(0,len(time_range)):
    freq = 0
    for num in time:
        if num<(count*100) and num>((count-1)*100):
            freq = freq + 1
    freq_list.append(freq)

freq_list
```

```python
plot_range = np.arange(len(time_range))
```

```python
plt.figure(figsize=(15,8))
plt.plot(plot_range, freq_list, lw=5, alpha=0.5, color='red')

plt.xticks(plot_range, time_range)

plt.xlabel("Time of Day",size=15); plt.ylabel("Frequency of Traffic Accidents",size=15)
plt.title("Time vs Accidents",size=20)
```
