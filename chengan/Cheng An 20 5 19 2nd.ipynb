```python
import numpy as np
import scipy.stats as sp
from matplotlib import pyplot as plt
```

```python
age = np.random.randint(20,80,80)
print(age)
```

```python
income = np.random.randint(20000,100000,80)

print(income)
```

```python
print(np.mean(age))
print(np.median(age))
print(np.percentile(age,50))
print(sp.mode(age))
```

```python
group1 = [] #20-30
group2 = [] #31-40
group3 = [] #41-50
group4 = [] #51-60
group5 = [] #>60
```

```python
for x in age:
    if 20<=x and x<=30:
        group1.append(x)
    elif 31<=x and x<=40:
        group2.append(x)
    elif 41<=x and x<=50:
        group3.append(x)
    elif 51<=x and x<=60:
        group4.append(x)
    else:
        group5.append(x)
print(group1)
print(group2)
print(group3)
print(group4)
print(group5)           
```

```python
group1.sort()
group2.sort()
group3.sort()
group4.sort()
group5.sort()

print(group1)
print(group2)
print(group3)
print(group4)
print(group5) 
```

```python
# Data to plot
labels = 'group1', 'group2', 'group3', 'group4', 'group5'
sizes = [len(group1), len(group2), len(group3), len(group4), len(group5)]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
explode = (0.1, 0.1, 0.1, 0.1, 0.2)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

```python
# Data to plot
labels = '20k-40k', '40k-50k', '50k-60k', '60k-80k', '>80k'
sizes = [len(g1), len(g2), len(g3), len(g4), len(g5)]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'green']
explode = (0, 0, 0, 0, 0.1)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

```python
# data to plot
n_groups = 5
age = (len(group1), len(group2), len(group3), len(group4), len(group5))
income = (len(g1), len(g2), len(g3), len(g4), len(g5))

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, age, bar_width,
alpha=opacity,
color='r',
label='Age')

rects2 = plt.bar(index + bar_width, income, bar_width,
alpha=opacity,
color='g',
label='Income')

plt.xlabel('Groups')
plt.ylabel('Frequency')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D','E'))
plt.legend()

plt.tight_layout()
plt.show()
```
