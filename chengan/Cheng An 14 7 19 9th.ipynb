```python
import scipy.stats as st
from fitter import Fitter
import numpy as np

from matplotlib import pyplot as plt

import pandas as pd
```

```python
data = pd.read_csv('basic statistic/basic statistic projects/brain_size.csv',sep=';')
data.head(39)
main_data = data['Weight']

data.info()
data.describe()
```

```python
main_data = data['Weight'].replace('.','1')
main_data = main_data.astype('int64')
main_data
```

```python
main_data.describe()
```

```python
main_data = main_data.replace(1,144)
main_data
```

```python
main_data.describe()
```

```python
fit_data = Fitter(main_data)
fit_data.fit()
fit_data.summary()
```

```python
fit_data.get_best()
```

```python
fit_data.hist()
```

```python
fit_data.plot_pdf(Nbest=4)
```
