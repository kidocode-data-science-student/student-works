```python
#functions
def morethan(z,locc,scalee):
    print(1 - st.norm.cdf(find_x(z,locc,scalee),loc=locc,scale=scalee),'x=',find_x(z,locc,scalee))

def lessthan(z,locc,scalee):
    print(st.norm.cdf(find_x(z,locc,scalee),loc=locc,scale=scalee),'x=',find_x(z,locc,scalee))

def middlethan(largerz,smallerz,locc,scalee):
    x1=find_x(largerz,locc,scalee)
    x2=find_x(smallerz,locc,scalee)
    print(st.norm.cdf(x1,loc=locc,scale=scalee)-st.norm.cdf(x2,loc=locc,scale=scalee),'x=',x1,'x2=',x2)
```

```python
morethan(1.247,8,15)
```
