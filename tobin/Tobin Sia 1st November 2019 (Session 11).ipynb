# Exponential Distribution

## 2. 1 PDF (Probability Density Function) of Exponential Distribution

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

x_values = range(1,40)
mean = np.mean(x_values)
lamb = 1 / mean
pdf_values = []
for i in x_values:
    pdf_values.append(lamb* math.exp(-lamb*i))
print(pdf_values)
plt.plot(x_values, pdf_values, color = "pink")
plt.show()
```

## 2. 2 CDF (Cumulative Distribution Function) of Exponential Distribution

```python
x_values = range(1,40)
mean = np.mean(x_values)
lamb = 1 / mean
cdf_values = []
for i in x_values:
    cdf_values.append(1 - math.exp((-lamb)*i))
print(cdf_values)
plt.plot(x_values, cdf_values, color = "purple")
plt.show()
```

```python
mean = 8
lamb = 1/mean
pdf_values = []
for i in x_values:
    pdf_values.append(lamb* math.exp(-lamb*i))
print(pdf_values)
plt.plot(x_values, pdf_values, color = "green")
```

Solving an Exponential Distribution using PYTHON (Numpy & Scipy)

Let's take the example above.

To plot the graph
Import all the libraries. matplolib, numpy and scipy
make a list with two different ways:
ranged list: np.arange(start_value, stop_value)
random number list: np.random.exponential(mean, scale of data) and sort() it

find f(x) by using st.expon.pdf(x_values, scale=1 / lambda)
plot your graph using matplotlib.pyplot

```python
import scipy.stats as st
mean = 8
lamb = 1.0 / mean

x_values = np.arange(1, 10)
expo_pdf = st.expon.pdf(x_values, scale=1/lamb)

plt.plot(x_values, expo_pdf, lw = 2, label = "PDF")
plt.legend(loc="best")
plt.show()
```
