### Computing Binomial

```python
import scipy.stats as st

x = 52
n = 100
p = 0.5
st.binom.pmf(x, n, p)
```

### Presenting Binomial

```python
import matplotlib.pyplot as plt
import scipy.stats as st

n = 100
p = 0.5

def P(X):
    return st.binom.pmf(X,n,p)

plt.bar([i for i in range(n)], [P(i) for i in range(n)], linewidth = 10, color = "darkcyan")

#Labels
plt.xlabel("Number of getting Heads after flipping a coin 100 times", fontsize = 12)
plt.ylabel("Probability of getting Heads", fontsize = 12)
plt.xticks(size = 10); plt.yticks(size = 10)

# consider the < 0.1 percent of the max probability as insignificant
l = []
for i in range(n):
    if P(i) > 0.001*P(p*200):
        l.append(i)

plt.xlim(l[0], l[-1])

plt.show()
```

### <font color = "darkcyan">Question 4</font>
