## Seaborn

### Dataset

```python
import pandas as pd  
import numpy as np

import matplotlib.pyplot as plt  
import seaborn as sns


D = sns.load_dataset('titanic')
D.head()
```

## Distributional plot

```python
sns.distplot(D['fare'])
# However y axis is kde spread
```

```python
#To remove Kernal density estimation

sns.distplot(D['fare'], kde = False)
```

```python
# To see details/ less details of graph 
sns.distplot(D['fare'] , kde=False , bins = 10)
```

```python
# Example 2
D['age'] = D['age'].fillna(np.mean(D['age']))

sns.distplot(D['age'] , kde = False)
```

```python
# Kde plot to observe the range of kde calculated
sns.kdeplot(D['age'])
```

### Joint plot

```python
sns.jointplot(x='age', y='fare', data=D)

# above the top x axis shows age distribution
# right of y axis shows fare distribution
# centre shows the spread
```

```python
sns.jointplot(x='age', y='fare', data=D, kind='kde')

#kind changes the type of plot
# scatter gives scatter plot
# reg gives the regular distributional plot alongside a regression of both x on y and y on x
# resid gives residual plot (show the relationship between an independent var and the response)
# kde shows the kde distribution (darker color higher kde)
# hex shows the spread along the x axis of the count
```

### Pair PLot (plot for boolean)

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

F = sb.load_dataset('iris')
sb.pairplot(F,hue = 'species')
plt.show()
```

```python
#Example 2
# IN pairplt, strings are filtererd but boolean are not, so remove boolean first otherwise error
F = D.drop(['alone','adult_male'] , 1)
print(F)
```

```python
F = F.dropna()
sb.pairplot(F , hue ="sex" )
```

## Categorical plot

### Bar plot (display standard values for each categorical collumn)

```python
#By default, display mean
sns.barplot(x='sex', y='age', data=D)
```

```python
# to display standard deviation
import numpy as np

import matplotlib.pyplot as plt  
import seaborn as sns

sns.barplot(x='sex', y='age', data=D, estimator=np.std)
```

```python
np.std(D['age'])
```

### Count plot (display counts for each categorical collumn)

```python
sns.countplot(x='sex', data=D)
```

### Box plot (display distribution for each categorical collumn in  terms of quartiles)

```python
sns.boxplot(x='sex', y='age', data=D)
```

```python
#adding another layer of category
sns.boxplot(x='sex', y='age', data=D, hue="survived")
```

### Vionlinplot (display distribution for y collumn with respect to x)

```python
sns.violinplot(x='sex', y='age', data=D , cut = 0 )

#symmetrical to individial centre line
# shown for males,there are a lot of males from age 20-40
```

```python
#adding another layer of category
sns.violinplot(x='sex', y='age', data=D, hue='survived')
```

```python
# for comparison, use split (very counter intuitive)
sns.violinplot(x='sex', y='age', data=D, hue='survived',split = True)
```

### Stripplot (display distribution for y collumn with respect to x)

```python
sns.stripplot(x='sex', y='age', data=D)
```

```python
# If there are overlapping values, 
# to see better, jitter takes random values from each collumn
sns.stripplot(x='sex', y='age', data=D, jitter=True)
```

```python
# add category
sns.stripplot(x='sex', y='age', data=D, jitter=True, hue='survived')
```

```python
# for split
sns.stripplot(x='sex', y='age', data=D, jitter=True, hue='survived', split=True)
```

### Swarmplot (mix of split and violin)

```python
sns.swarmplot(x='sex', y='age', data=D)
```

```python
# add category
sns.swarmplot(x='sex', y='age', data=D, hue='survived' , split = True)
```

```python
#Combining Swarm and Violin
sns.violinplot(x='sex', y='age', data=D)  
sns.swarmplot(x='sex', y='age', data=D, color='White')
```

```python
# for hue
sns.violinplot(x='sex', y='age', data=D , hue = 'survived' , split = True)  
sns.swarmplot(x='sex', y='age', data=D, hue = 'survived' , color='White' , split = 'True')
```
