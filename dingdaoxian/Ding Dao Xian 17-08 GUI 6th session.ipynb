```python
from ipywidgets import interact, interactive ,fixed
import ipywidgets as widgets
from IPython.display import display

#database
l = []
l2 =[]
l3 =[]
l4 = []

#Display
disp = widgets.FloatText(description='value')


#Numbers
z = 0
def NUME(k):
    global z 
    z = z*10 + int(k.description)
    disp.value = z

one = widgets.Button(description = '1')
one.on_click(NUME)
two = widgets.Button(description = '2')
two.on_click(NUME)
three = widgets.Button(description = '3')
three.on_click(NUME)
four = widgets.Button(description = '4')
four.on_click(NUME)
five = widgets.Button(description = '5')
five.on_click(NUME)
six = widgets.Button(description = '6')
six.on_click(NUME)
seven = widgets.Button(description = '7')
seven.on_click(NUME)
eight = widgets.Button(description = '8')
eight.on_click(NUME)
nine = widgets.Button(description = '9')
nine.on_click(NUME)
zero = widgets.Button(description = '0')
zero.on_click(NUME)

#reset
r = widgets.Button(description = 'r')

def reset(args):
    global z
    z = 0
    l.clear()
    l2.clear()
    l3.clear()
    l4.clear()
    disp.value = z

r.on_click(reset)


#add
ad = widgets.Button(description = "+")

def add(args):
    global z,p
    l.append(z)
    z = 0

ad.on_click(add)

#subtract
minu = widgets.Button(description = "-")

def minus(args):
    global z
    l2.append(z)
    z = 0

minu.on_click(minus)

#multiply
mul = widgets.Button(description = "x")

def multiply(args):
    global z
    l3.append(z)
    z = 0

mul.on_click(multiply)

#division
div = widgets.Button(description = "/")

def divide(args):
    global z
    l4.append(z)
    z = 0

div.on_click(divide)

#equal
eq = widgets.Button(description = "=")

def equal(args):
    global z
    if len(l4)>0:
        for i in l4:
            z = i/z
        disp.value = z
    if len(l3)> 0:
        k = 1
        for i in l3:
            k = k*i
        z = z*k
        l3.clear()
        disp.value = z
    if len(l2) > 0:
        z = l2[0]*2-sum(l2)-z
        l2.clear()
        disp.value = z
    if len(l) > 0 :
        z = z + sum(l)
        l.clear()
        disp.value = z

eq.on_click(equal)


#Resultant
display(disp)
display(widgets.HBox((one,two,three , ad)))
display(widgets.HBox((four,five,six, minu)))
display(widgets.HBox((seven,eight,nine, mul)))
display(widgets.HBox((zero,r,eq, div)))
```
