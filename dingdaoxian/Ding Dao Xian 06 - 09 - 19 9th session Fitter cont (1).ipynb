```python
import pandas as pd
v = pd.read_csv("500_Person_Gender_Height_Weight_Index.csv")
v.rename(columns={'Height' : 'Height(cm)' , "Weight" :"Weight(pounds)"}, inplace = True)
v.insert(2,'Height(m)',v['Height(cm)']/100)
v.insert(3,'Weight(kg)',v['Weight(pounds)']*0.453592)
v.insert(5,'BMI',v['Weight(kg)']/v['Height(m)']**2)
v
```

```python
print(v.describe())
```

# Questions

Question1. Find which distribution fits best for "BMI" data.(Limit Distributions to only Uniform, Normal and Exponential)

```python
from fitter import Fitter

f = Fitter(v['BMI'])
f.fit()
```

```python
f.summary()
```

Question2. Find the Parameters of the distribution that you chose in question 1.

```python
param = f.fitted_param['nakagami']
param
```

Question3. Plot the pdf of the distribution that you chose in Question1.

```python
from pylab import linspace, plot 
import scipy.stats

X = linspace(min(v['BMI']),max(v['BMI']), len(v['BMI']))
naka_values = scipy.stats.nakagami.pdf(X, *param)
plot(X,naka_values,'-r')
```

# Fitter Project

```python
import pandas as pd
df=pd.read_csv('monthly_population_estimatesNA-EST2018-01.csv',header=2).dropna()
df.head()
```

```python
for i in df['Resident Population']:
    p = i.replace(",","")
    df['Resident Population'].replace(to_replace=[i], value = int(p), inplace = True)
df['Resident Population']
```

## Finding the best fitted Distribution 



```python
from fitter import Fitter

H = Fitter(df['Resident Population'])
H.fit()
```

```python
H.summary()
```

```python
param3 = H.fitted_param['beta']
param3
```

## Prediction

```python
from pylab import linspace, plot 
import scipy.stats

X = linspace(min(df['Resident Population']),max(df['Resident Population']), len(df['Resident Population']))
beta_values = scipy.stats.beta.pdf(X, *param3)
plot(X,beta_values)
```
