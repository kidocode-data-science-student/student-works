## <b>How to create a graph using Python</b>

```python
import numpy as np
from scipy.stats import poisson
from matplotlib import pyplot as plt

dist = poisson(mu=1.6)
x = np.arange(11)

plt.plot(x, dist.pmf(x),drawstyle = 'steps-mid')
plt.show()
```

```python
import numpy as np
from scipy.stats import poisson
from matplotlib import pyplot as plt
from matplotlib import ticker as ticker


mu_values = [1.6, 4.8] 

linestyles = ['-.', '-'] 
colors = ['green', 'yellow'] 
c = ['red' , 'black']

fig, ax = plt.subplots(figsize=(10, 7))

for mu, linestyles, colors , c  in zip(mu_values, linestyles, colors , c):
    # For loop to create a poisson distribution
    dist = poisson(mu)
    x = np.arange(0, 11.0)

    plt.plot(x, dist.pmf(x), ls = linestyles, color= colors, label='{}'.format(mu), drawstyle = 'steps-mid')
    plt.vlines(x, 0, dist.pmf(x), color = c , lw=1, alpha=0.5) #lw : linewidth , alpha : opacity of line


plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.yticks(np.arange(min(x), max(x)+1, 0.05)) 
#ticks : spacing between consequetive numbers in each axis

ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.2f'))
# shows 2 decimal place , similar to pxl in pie chart

plt.xlim(0, 11) 
plt.ylim(0, 0.4)

plt.xlabel('$x$') 
plt.ylabel(r'$p(x|\mu)$')
plt.title('Poisson Distribution') 

plt.legend()
plt.show()
```

### CDF

```python
import numpy as np
from scipy.stats import poisson
from matplotlib import pyplot as plt
from matplotlib import ticker as ticker


mu_values = [1.6, 4.8] 
linestyles = ['-.', '-'] 
colors = ['blue', 'red']


fig, ax = plt.subplots(figsize=(10, 7))

for mu, linestyles, colors in zip(mu_values, linestyles, colors):
    dist = poisson(mu)
    x = np.arange(0, 11.0)
    plt.plot(x, dist.cdf(x), ls = linestyles, color= colors, label=r'$\mu=%0.1f$' % mu, drawstyle = 'steps-mid')
    #The change is over here at the plt.plot (dist.cdf)

    
    
plt.xticks(np.arange(min(x), max(x)+1, 1.0))
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.2f'))
plt.yticks(np.arange(min(x), max(x)+1, 0.05)) 
    
plt.xlim(0, 11) 
plt.ylim(0, 1.0)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\mu)$')
plt.title('Poisson Distribution') 

plt.legend()
plt.show()
```

### Sample poisson simulation

```python
import numpy as np
import matplotlib.pyplot as plt

s = np.random.poisson(4.8,1000)
count, bins, ignored = plt.hist(s,12,density = True)
plt.show()
```
