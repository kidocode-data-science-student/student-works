# Data visualisation with matplotlib and Seaborn

## Matplotlib Line graph

```python
#Basic Syntax

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,9, 20)
y = x **3

plt.plot(x, y, 'r') #the r stands for red, if "ro" or "bo", it will be circle plots
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Cube Function')
plt.show()
```

```python
#two numpy arrays x and y
#1stly , use linespace to generate 20 numbers between -10 to 9
#2nd, define the variable y
#3rd , use plot to plot the lines
#more numbers generated through nplinespace, the more accurate the graph
```

###  Multiple Line graph in a canvas

```python
#subplot : to make more than one plots on one canvas
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 9, 20)

y = x ** 3

plt.subplot(2,2,1) # plt.subplot(number of rows,number of collumns, position the plot takes)
plt.plot(x, y, 'b*-')
plt.subplot(2,2,2)
plt.plot(x, y, 'y--')
plt.subplot(2,2,3)
plt.plot(x, y, 'b*-')
plt.subplot(2,2,4)
plt.plot(x, y, 'y--')
```

### Ploting in object oriented way

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 9, 20)

y = x ** 3

figure = plt.figure() #the plt.figure returns a "general canvas" to manipulate

axes = figure.add_axes([12, 0.2,0.8,0.8]) #.add_axes([starting x, starting y , width, height]) all relative to size of screen
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 9, 20)

y = x ** 3

z = x ** 2

figure = plt.figure()

axes = figure.add_axes([0.0, 0.0, 1.5, 1.5])
axes2 = figure.add_axes([0.2, 1, 0.3, 0.3]) 
#inset 2 different axes and are manipulated in the canvas

axes.plot(x, y, 'b')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Cube function')

axes2.plot(x, z, 'r')
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('Square function')
```

### Quick Subploting if axis stays the same

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 9, 20)

y = x ** 3

z = x ** 2

fig, axes = plt.subplots(nrows=2, ncols=3,squeeze = True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 9, 20)

z = x ** 2

figure, axes = plt.subplots(nrows=2, ncols=3, sharex='row')

for rows in axes:
    for ax1 in rows:
        ax1.plot(x, z, 'b')
        ax1.set_xlabel('X - axis')
        ax1.set_ylabel('Y - axis')
        ax1.set_title('Square Function')
```

### FigSize

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 9, 20)

y = x ** 3

z = x ** 2

figure, axes = plt.subplots(figsize = (6,8)) #means x = 6inch , y = 8inch

axes.plot(x, z, 'r')
axes.set_xlabel('X-Axis')
axes.set_ylabel('Y-Axis')
axes.set_title('Square Function')
```

### Legend

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 9, 20)

y = x ** 3

z = x ** 2

figure = plt.figure()

axes = figure.add_axes([0,0,1,1])

axes.plot(x, z, label="Square Function")
axes.plot(x, y, label="Cube Function")
axes.legend(loc = 3)  #loc changes location of the legend it can be strings such as "upper left"

#As long as there's a label variable and legend is placed under it then thew lave will appear
```

### Color option

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 9, 20)

y = x ** 3

z = x ** 2

figure = plt.figure()

axes = figure.add_axes([0,0,1,1])

axes.plot(x, z, "r" ,label="Square Function") #full letters can be written  but the initials of the coolors ussually has a value
axes.plot(x, y, "g", label="Cube Function")
axes.legend(loc=4)

#by default, different plots on same page will have different colors
```

## Matplotlib stackplot

```python
#Stack plot is an extension of bar chart or line chart 
#which breaks down data from different categories and stack them together
#so that comparison between the values from different categories can easily be made.
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]

y = np.vstack([y1, y2, y3])

labels = ["Fibonacci ", "Evens", "Odds"]


#Sinch stackplots are multiple plots, they are technically a subplot
fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc='upper left')
plt.show()

fig, ax = plt.subplots()
ax.stackplot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

year = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]

player1 = [8,10,17,15,23,18,24,29]
player2 = [10,14,19,16,25,20,26,32]
player3 = [12,17,21,19,26,22,28,35]


fig, ax = plt.subplots()
ax.stackplot(year, player1, player2, player3, colors = ['y','r','b'], labels = ["player1","player2","player3"])
ax.legend()

plt.title('Goals by three players')
plt.xlabel('year')
plt.ylabel('Goals')
plt.show()
```
